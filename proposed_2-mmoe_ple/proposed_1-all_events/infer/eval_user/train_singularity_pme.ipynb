## Resigter Component

```python
from azure.ai.ml.entities import CommandComponent, PipelineComponent

component_inputs = {}
for k, v in command_job.inputs.items():
    if v.type=='string':
        component_inputs[k] = Input(type="string", default=v._original_data)
    else:
        component_inputs[k] = Input(type=v.type)
print(component_inputs)

component = CommandComponent(
    name="inference_eval_user",
    version="3.5",
    code=command_job.code,
    command=command_job.command,
    environment=command_job.environment,
    inputs=component_inputs,
    outputs={k: Output(type=v.type) for k, v in command_job.outputs.items()},
    resources=command_job.resources,
    distribution=command_job.distribution
)
print(component.outputs)
ml_client.components.create_or_update(component)
```
